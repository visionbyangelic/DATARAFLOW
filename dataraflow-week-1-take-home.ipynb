````python
#Write a Python program to print `"Hello, World!"`

print('Hello, World!')
````

````python
#. Create a program that takes a user’s name and age as input, then prints:  `"Hello [name], you are [age] years old!"`.

user_name = input("Input your name: ")
user_age = input("Input your age: ")

print("Hello " + user_name + ", you are " + user_age + " years old!")
````

````python
# 3. Write a program to check whether a number entered by the user is even or odd.

user_number = int(input("Input a number: "))
if user_number % 2 == 0:
    print(str(user_number) + " is even")
````

````python
# 4. Create a program that prints the multiplication table of a number (from 1–12).

user_num= int(input("Input a number: "))
print("\nHere is the multiplication table for", user_num, ":\n")
for i in range(1, 13):
    result = user_num * i
    print(user_num, "x", i, "=", result)
````

````python
# 5. Write a program that calculates the factorial of a number using a loop.


number = int(input("Enter a number to find its factorial: "))
factorial = 1

for i in range(1, number + 1):
    factorial = factorial * i   

print("The factorial of", number, "is", factorial)
````

````python
#Write a program that finds the largest number in a given list.

numbers = [3, 7, 2, 9, 5, 25, 27, 4, 8]
largest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num

print("The largest number is:", largest)
````

````python
#Create a Python program that counts how many vowels are in a given string.

string = 'Angelic charles is so pretty'
vowels = ('a', 'e', 'i', 'o', 'u')
count = 0

for char in string.lower():  # .lower() makes it case-insensitive
    if char in vowels:
        count += 1
````

````python
#Write a function that takes two numbers and returns their sum, difference, and product.

def calculate(a, b):
    s = a + b
    d = a - b
    p = a * b
    return s, d, p

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

sum_result, diff_result, prod_result = calculate(num1, num2)

print("Sum:", sum_result)
print("Difference:", diff_result)
print("Product:", prod_result)
````

````python
#Create a program that reverses a string without using slicing ([::-1]).

text = input("Enter a string: ")
reversed_text = "".join(reversed(text))
print("Reversed string:", reversed_text)
````

````python
#Write a program that generates the first 10 numbers of the Fibonacci sequence.

a, b = 0, 1

print("First 10 Fibonacci numbers:")

for _ in range(10):
    print(a, end=" ")
    a, b = b, a + b
````

````python
#Student Grades System
#Write a program to accept marks for 5 subjects, calculate the average, and display the grade based on:
#A: 80–100, B: 60–79, C: 40–59, F: < 40

m1 = int(input("Enter marks for subject 1: "))
m2 = int(input("Enter marks for subject 2: "))
m3 = int(input("Enter marks for subject 3: "))
m4 = int(input("Enter marks for subject 4: "))
m5 = int(input("Enter marks for subject 5: "))

avg = (m1 + m2 + m3 + m4 + m5) / 5

if avg >= 80:
    grade = "A"
elif avg >= 60:
    grade = "B"
elif avg >= 40:
    grade = "C"
else:
    grade = "F"

print("Average is:", avg)
print("Grade is:", grade)

````

````python
#Simple Calculator
#Build a calculator program that can perform addition, subtraction, multiplication, and division based on user input.

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
op = input("Enter operation + - * / : ")

if op == "+":
    print("Answer is", a+b)
elif op == "-":
    print("Answer is", a-b)
elif op == "*":
    print("Answer is", a*b)
elif op == "/":
    if b != 0:
        print("Answer is", a/b)
    else:
        print("cannot divide by zero")
else:
    print("wrong operation")
````

````python
# Shopping Cart
#Create a program where a user can add items to a shopping cart (dictionary with item names and prices), then display the total cost.

cart = {}
done = False

while not done:
    item = input("Enter item name (or done): ")
    if item == "done":
        done = True
    else:
        price = float(input("Enter price: "))
        cart[item] = price

total = 0
for i in cart:
    total = total + cart[i]

print("Cart:", cart)
print("Total is", total)
````

````python
#Guessing Game
#Write a program where the computer randomly selects a number between 1–20, and the user must guess it. Give hints if the guess is too high or too low.

import random

num = random.randint(1, 20)

guess = 0
while guess != num:
    guess = int(input("Guess number between 1 and 20: "))
    print("You guessed:", guess)   # show user’s input
    if guess < num:
        print("Too low")
    elif guess > num:
        print("Too high")

print("Correct! The number was:", num)   # show correct number
````

````python
#File Handling
#Write a program that accepts a list of student names from the user, saves them in a text file, and then reads and displays them.

names = []
done = False

while not done:
    n = input("Enter name (or done): ")
    if n == "done":
        done = True
    else:
        names.append(n)

f = open("students.txt", "w")
for n in names:
    f.write(n + "\n")
f.close()

print("Names saved. Now reading file:")

f = open("students.txt", "r")
for line in f:
    print(line.strip())
f.close()
````
